```python
import sys
sys.path.append('..')
from getData import getProducts
from numpy import mean
import matplotlib.pyplot as plt

def plotData():
    products = getProducts(time=True)
    allScrapedDates = {}
    for i in products.values():
        for _, price, date in i:
            if date not in allScrapedDates.keys():
                allScrapedDates[date] = []
            allScrapedDates[date].append(price)
    for dateTime in allScrapedDates:
        allScrapedDates[dateTime] = float(mean(allScrapedDates[dateTime]))
    plt.plot(list(allScrapedDates.keys()), list(allScrapedDates.values()))
    plt.xticks(rotation=45)

plotData()
```
