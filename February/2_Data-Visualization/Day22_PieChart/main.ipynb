```python
import sys
sys.path.append('..')
from getData import getProducts
import matplotlib.pyplot as plt

def plotProducts():
    products = getProducts()
    categories: list[str] = []
    amountOfProducts: list[int] = []
    for category in products:
        categories.append(category)
        amountOfProducts.append(len(products[category]))
    plt.pie(amountOfProducts, labels=categories)
    plt.show()

plotProducts()
```
