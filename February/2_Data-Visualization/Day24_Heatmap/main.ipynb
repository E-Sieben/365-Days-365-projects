```python
import sys
sys.path.append('..')
from getData import getProducts
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
def plotData():
    products_by_category = getProducts()

    data = []
    for category, products in products_by_category.items():
        for product, price in products:
            data.append({'Category': category, 'Product': product, 'Price': price})
    df = pd.DataFrame(data)
    price_bins = [0, 10, 20, 50, 100, 200, 500, 1000, float('inf')]
    price_labels = ['0-10', '10-20', '20-50', '50-100', '100-200', '200-500', '500-1000', '1000+']

    df['Price Range'] = pd.cut(df['Price'], bins=price_bins, labels=price_labels)
    heatmap_data = pd.crosstab(df['Category'], df['Price Range'])
    sns.heatmap(heatmap_data,cmap='BuPu')
    plt.ylabel('Category')
    plt.xlabel('Price Range')
    plt.tight_layout()
    plt.show()

plotData()
```
